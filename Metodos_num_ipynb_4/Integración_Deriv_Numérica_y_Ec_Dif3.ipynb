# **Ecuaciones diferenciales**

$$y'=ycos^2(x)+x+1.1,  y(0)=2;  0 \leq x \leq 2$$

```python
import matplotlib.pyplot as plt
import numpy as np

def fun(x,y):
  return y*(np.cos(x))**2+x+1.1
```

### Runge Kutta primer orden.

$$y_{i+1}=y_{i}+f(x_{i},y_{i})*h $$

```python
def rungekutta1(f,a,b,h,y0):
  x=[a]
  y=[y0]
  while x[-1]<b:
    y.append(y[-1]+f(x[-1],y[-1])*h)
    x.append(x[-1]+h)
    return x,y
```

### Runge Kutta orden 2

```python
def rungekutta2(f,a,b,h,y0):
  x=[a]
  y=[y0]
  while x[-1]<b:
    k1=f(x[-1],y[-1])
    k2=f(x[-1]+3/4*h,y[-1]+3/4*k1*h)
    y.append(y[-1]+(1/3*k1+2/3*k2)*h)
    x.append(x[-1]+h)
  return x,y
```

### Runge Kutta 4 orden

```python
def rungekutta4(f,a,b,h,y0):
  x=[]
  y=[]
  x.append(a)
  y.append(y0)
  while x[-1]<b:
    k1=f(x[-1],y[-1])
    k2=f(x[-1]+0.5*h,y[-1]+0.5*k1*h)
    k3=f(x[-1]+0.5*h,y[-1]+0.5*k2*h)
    k4=f(x[-1]+h,y[-1]+k3*h)
    y.append(y[-1]+(h/6)*(k1+2*k2+2*k3+k4))
    x.append(x[-1]+h)
  return x, y
```

```python
h=0.5; a=0; b=2; y0=2
print("TamaÃ±o de paso (h)= ",h)
x2,y2=rungekutta2(fun,a,b,h,y0)
x1,y1=rungekutta1(fun,a,b,h,y0)
x4,y4=rungekutta4(fun,a,b,h,y0)

# Graficas:
plt.figure(figsize=(16, 10), dpi=80) # tamaÃ±o grÃ¡fica
plt.plot(x1,y1,label='RK primer orden')
plt.plot(x2,y2,label='RK segundo orden')
plt.plot(x4,y4,label='RK cuarto orden')
plt.legend()
plt.show()
```

```python
h=0.1; a=0; b=2; y0=2
print("TamaÃ±o de paso (h)= ",h)
x2,y2=rungekutta2(fun,a,b,h,y0)
x1,y1=rungekutta1(fun,a,b,h,y0)
x4,y4=rungekutta4(fun,a,b,h,y0)

# Graficas:
plt.figure(figsize=(16, 10), dpi=80) # tamaÃ±o grÃ¡fica
plt.plot(x1,y1,label='RK primer orden')
plt.plot(x2,y2,label='RK segundo orden')
plt.plot(x4,y4,label='RK cuarto orden')
plt.legend()
plt.show()
```

```python
h=0.05; a=0; b=2; y0=2
print("TamaÃ±o de paso (h)= ",h)
x2,y2=rungekutta2(fun,a,b,h,y0)
x1,y1=rungekutta1(fun,a,b,h,y0)
x4,y4=rungekutta4(fun,a,b,h,y0)

# Graficas:
plt.figure(figsize=(16, 10), dpi=80) # tamaÃ±o grÃ¡fica
plt.plot(x1,y1,label='RK primer orden')
plt.plot(x2,y2,label='RK segundo orden')
plt.plot(x4,y4,label='RK cuarto orden')
plt.legend()
plt.show()
```
